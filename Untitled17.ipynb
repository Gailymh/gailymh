```python
import numpy as np #linear algebra
import pandas as pd #data proccessing
import matplotlib.pyplot as plt #to visualize data features
import seaborn as sns #to show the correlation between features
```

```python
#loading of data
```

```python
data=pd.read_csv('/content/Housing_californiya.csv')
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
#splitting of data 
```

```python
x=data.drop(['median_house_value'],axis=1)
y=data['median_house_value']
```

```python
x
```

```python
y
```

```python
#preprocessing the data and train test split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
#0.2 percent of the data will be used for evaluation
#training data
```

```python
train_data=x_train.join(y_train)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

x_train,y_train=train_data.drop(['median_house_value'],axis=1),train_data['median_house_value']
x_train_s=scaler.fit_transform(x_train)

reg=LinearRegression()

reg.fit(x_train_s,y_train)
```
